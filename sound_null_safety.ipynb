### Sound Null Safety   

->  Making any type nullable  

        String? name = null;
        print(name);   
        name = "emre";   
        print(name);   

        output:  
        flutter : null  
        flutter : emre  


        int? age = 20;  
        age = null;  


        List<String?>? names = ['emre','ozan',null];  
        names = null;  


->  Cherry-picking non-null values  

        const String? firstName = "null";  
        const String? middleName = "null";  
        const String? lastName = "parlak";  

        if (firstName != null){  
            print("first name is the first non-null value");  
        }else if (middleName != null){  
            print("middle name is the first non-null value");  
        }else if (lastName != null){  
            print("last name is the first non-null value");  
        }  

        const firstNonNullValue = firstName ?? middleName ?? lastName;  


->  Null-aware assignment operator  

        String? name = firstName;  
        name ??= middleName;  
        name ??= lastName;  
        print(name);  

        output :  

        flutter : parlak  


->  Conditional invocation  


        void test(List<String>? names){    #it gives you an error if you do not check whether it can be null, so that you need to add an if condition to tell flutter that i check whether it is null.  
            if(names != null){  
                final length = names.length;  
            }  else {  
                length = 0;  
            }  
        }  

        #use this instead  

        void test(List<String>? names){  
            final length = names?.length ?? 0;  
        }
