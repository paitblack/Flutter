        void test(AnimalType animalType){  
            switch (animalType) {  
                case AnimalType.rabbit:  
                    print("rabbit");  
                    return;  
                case AnimalType.cat:  
                    print("cat");  
                    return;  
                case AnimalType.dog:  
                    print("dog")  
                    return;  
            }  
        }  

        test(AnimalType.cat);  # output : cat

### Factory Constructors  

Can return instances that are not of the same class.  

Key features :    
No New Object Needed -> Can return an existing object instead of creating a new one.  
Encapsulation -> Can hide complex logic for object creation.  
Flexibility -> Can return objects of the same class or different classes.  
Optimization -> Helps save memory and improve performance.  

        class Cat{  
            final String name;  
            Cat(this.name);  
            factory Cat.miyav(){  
                return Cat("miyav");  
            }  
        }  


### Custom Operators  

        class Cat{  
            final String name;  
            Cat(this.name);  

            @override  
            bool operator == (covariant Cat other) => other.name == name;  

            @override  
            int get hashCode => name.hashCode;  
        }
